```python
import importlib
import utility
importlib.reload(utility)

model = pm.Model()

parameters = { k: utility.log_normal_params_from_ci(**v) for k, v in intervals.items()}
log_normals = utility.log_normalify_parameters(model, parameters);
```

```python
import cash
importlib.reload(cash)

with model:
    pm.Deterministic("value_per_dollar_transferred", cash.cash_transfers(**log_normals))
```

```python
with model:
    trace = pm.sample(2000)
```
